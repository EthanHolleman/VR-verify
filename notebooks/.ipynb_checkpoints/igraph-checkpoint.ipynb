```python
%%R
library(igraph)
library(ggraph)
```

Read in example blast alignment.

```python
%%R

f <- '../output/blast-results/with-headers/all-results-cat.tsv'
df <- as.data.frame(read.table(f, sep='\t', header=TRUE))
df <- subset(df, length > 250 & pident > 95)
head(df)
```

```python
%%R

df.graph <- df[, c('qseqid', 'description', 'length', 'mismatch', 'pident')]
graph <- graph_from_data_frame(df.graph, directed = TRUE)

# edit color
V(graph )$color <- ifelse(V(graph)$name %in% df$qseqid, "lightblue", "orange")
V(graph)$label.color <- 'black'
#V(graph)$size <- 25
E(graph)$color <- 'black'

l <- layout.bipartite(graph, types=V(graph)$name %in% df$qseqid)

#visual_style["layout"] = l
#visual_style["bbox"] = (300, 300)
#visual_style["margin"] = 10
plot(graph, layout = l[, c(2,1)], color='skyblue')
```

```python
%%R

library(ggpubr)
```

```python
%%R -w 5 -h 3 --units in -r 200
ggraph(graph) + 
    geom_node_point(aes(color=color)) + 
    geom_edge_link(width=0.1) + 
    geom_node_text(label=V(graph)$name, size=2, check_overlap=TRUE) +
    theme_pubr()
```
